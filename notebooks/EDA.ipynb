# Análisis Exploratorio de Datos (EDA) - Anuncios de Vehículos

Este notebook realiza un análisis exploratorio básico del conjunto de datos `vehicles_us.csv` utilizando pandas y plotly-express.

```python
import pandas as pd
import plotly.express as px
```

## Carga y exploración básica de los datos

```python
# Intentar cargar desde el directorio actual o el directorio padre
try:
    df = pd.read_csv('vehicles_us.csv')
except FileNotFoundError:
    df = pd.read_csv('../vehicles_us.csv')

df.head()
```

```python
df.info()
```

## Visualizaciones con Plotly Express

### 1. Histograma del Odómetro (Kilometraje)
Visualizamos cómo se distribuye el kilometraje de los vehículos en venta.

```python
fig_hist = px.histogram(df, x='odometer', title='Distribución del Kilometraje de Vehículos')
fig_hist.show()
```

### 2. Gráfico de Dispersión (Precio vs Odómetro)
Analizamos la relación entre el precio del vehículo y su kilometraje.

```python
fig_scatter = px.scatter(df, x='odometer', y='price', title='Precio vs Kilometraje')
fig_scatter.show()
```
